```python
#Selain menyertakan dua buah deret bilangan sebagai parameter untuk sumbu x dan y, terdapat parameter ketiga yang bisa kita sertakan untuk mengatur warna dan jenis plotting.
#Pengaturan parameter ketiga ini sepenuhnya mengadopsi gaya formatting pada MATLAB.
#Nilai default untuk parameter ketiga ini adalah '-b'. b berarti blue, - berarti garis.
#Pemanggilan fungsi axis() menyertakan sebuah list [xmin, xmax, ymin, ymax].
plt.plot([1,2,3,4],[1,4,9,16],'ro') #r berarti merah, o berarti dot. #jika ingin biru pilih b, dan g untuk hijau. #jika marker ingin diganti bintang sertaka *.
plt.axis([0,6,0,20])
plt.show()
```

```python
#Multi plot dalam satu pemanggilan fungsi plot()

t= np.arange(0.,5.,0.2)
t
```

```python
plt.plot(t,t,'r--') #warna merah , garis
plt.plot(t,t**2,'bs') #warna biru , kotak/square
plt.plot(t,t**3,'g^') #warna hijau, caping
plt.show()
```

```python
#merupakan cara sederhana dari sebelumnya
plt.plot(t,t,'r--',
         t,t**2,'bs',
         t,t**3,'g^')
plt.show()
```

```python
#Matplotlib juga memungkinkan kita untuk melakukan plotting sekumpulan nilai yang tersimpan dalam struktur data yang disertai keywords argument (kwargs), seperti pada dictionary dan Pandas DataFrame.

data={'a': np.arange(50), 'c': np.random.randint(0,50,50), 'd': np.random.randn(50)}
data['b'] = data['a'] + 10*np.random.randn(50)
data['d'] = np.abs(data['d'])*100

data
```

```python
plt.scatter('a','b', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
plt.scatter('a','b', c='c',s='d',data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```
